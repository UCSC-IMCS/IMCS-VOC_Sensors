```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sensor_data.csv')

# Convert the 'Timestamp' column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Define colors for each line
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'black', 'gray']

# Plotting
plt.figure(figsize=(10, 6))

# Plot each column against the timestamp with different colors and labels
for i, column in enumerate(df.columns[1:]):
    plt.plot(df['Timestamp'], df[column], color=colors[i], label=column)

# Set labels and title
plt.xlabel('Time')
plt.ylabel('Analog Values')
plt.title('VOC Sensor Data Readings Over Time')
plt.legend()
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
